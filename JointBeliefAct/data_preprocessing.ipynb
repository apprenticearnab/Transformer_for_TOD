**Tokenizing Training data**

```python
from tokenizers import ByteLevelBPETokenizer
tokenizer = ByteLevelBPETokenizer()
#tokenizer.train(["belief_train.txt","context_train.txt"], vocab_size=30000 ,special_tokens=["<PAD>","<|belief|> "," <|endofbelief|>","<|context|>"," <|user|> "," <|system|> ","<|endofcontext|>"])
tokenizer.train(["Processed_data/simptod/belief_train.txt","Processed_data/simptod/context_train.txt","Processed_data/simptod/action_train.txt"], vocab_size=30000 )
```

```python
tokenizer.add_special_tokens(["<PAD>","<|belief|>"," <|endofbelief|>","<|context|>"," <|user|> "," <|system|> ","<|endofcontext|>","<|action|>"," <|endofaction|>"])
```

```python
tokenizer.save_model('.')
```

```python
f=open('Processed_data/simptod/belief_train.txt','r',errors = 'ignore')
raw=f.read()
#raw=raw.lower()
output = raw.split("\n")
f1=open('Processed_data/simptod/context_train.txt','r',errors = 'ignore')
raw=f1.read()
#raw=raw.lower()
input = raw.split("\n")
```

```python
print(input[1])
print(output[1])
```

```python
tokenized_input = []
for i in range(len(input)):
  tokenized_input.append(tokenizer.encode(input[i]))
tokenized_output = []
for i in range(len(output)):
  tokenized_output.append(tokenizer.encode(output[i]))
```

```python
vocab_dict = tokenizer.get_vocab()
inv_vocab_dict = {v: k for (k, v) in vocab_dict.items()}
```

**Converting ids to tokens**

```python
def convert_ids_to_tokens(id_list):
  tokens = []
  for ids in id_list:
    tokens.append(inv_vocab_dict[ids])
  return tokens
```

```python
print(tokenized_input[1].ids)
print(tokenized_output[1].ids)
print(convert_ids_to_tokens(tokenized_input[1].ids))
print(convert_ids_to_tokens(tokenized_output[1].ids))
```

```python
decoded_input = []
for i in range(len(input)):
  decoded_input.append(tokenizer.decode(tokenized_input[i].ids,skip_special_tokens=False))
decoded_output = []
for i in range(len(output)):
  decoded_output.append(tokenizer.decode(tokenized_output[i].ids,skip_special_tokens=False))
```

```python
print(decoded_input[1])
print(decoded_output[1])
```
