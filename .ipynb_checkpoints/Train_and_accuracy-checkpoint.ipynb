```python
epochs = []
loss_plot = []
accuracy = []
def run_epoch(data_iter, model, loss_compute):
    "Standard Training and Logging Function"
    start = time.time()
    total_tokens = 0
    total_loss = 0
    tokens = 0
    loss1 = []
    index = []
    for i, batch in enumerate(data_iter):
        reference = []
        candidate = []
        out = model.forward(batch.src.to(device), batch.trg.to(device), 
                            batch.src_mask.to(device), batch.trg_mask.to(device))
        loss = loss_compute(out, batch.trg_y.to(device), batch.ntokens.to(device))
        loss1.append(loss)
        index.append(i)
        total_loss += loss
        total_tokens += batch.ntokens
        tokens += batch.ntokens
        if i % 200 == 1:
            elapsed = time.time() - start
            max , idx = torch.max(model.generator(out),dim=2)
            a = idx.tolist()
            #print(a)
            b = batch.trg_y.tolist()
            print(idx.size())
            print(batch.trg_y.size())
            for id in a:
                reference.append(tokenizer.decode(id,skip_special_tokens=True).split(','))
            for p in batch.trg_y.tolist():
                candidate.append(tokenizer.decode(p,skip_special_tokens=True).split(',')) 
            print(reference)
            print(candidate)
            accuracy = joint_accuracy(candidate,reference,d_c = False,type2_c = False)
            slt_accuracy = slot_accuracy(candidate,reference)
            print("Epoch Step: %d Loss: %f Tokens per Sec: %f Joint Accuracy: %f Slot Accuracy: %f" %
                    (i, loss / batch.ntokens, tokens / elapsed,accuracy,slt_accuracy))#,slt_accuracy))
            start = time.time()
            tokens = 0
    return total_loss / total_tokens,accuracy,slt_accuracy
```

```python
class SimpleLossCompute:
    "A simple loss compute and train function."
    def __init__(self, generator, criterion, opt=None):
        self.generator = generator
        self.criterion = criterion
        self.opt = opt
        
    def __call__(self, x, y, norm):
        x = self.generator(x)
        loss = self.criterion(x.contiguous().view(-1, x.size(-1)), 
                              y.contiguous().view(-1)) / norm
        loss.backward()
        if self.opt is not None:
            self.opt.step()
            self.opt.optimizer.zero_grad()
        return loss.item() * norm
```

```python
criterion = LabelSmoothing(size=V, padding_idx=0, smoothing=0.0)
model_opt = NoamOpt(model.src_embed[0].d_model, 1, 400,
        torch.optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.98), eps=1e-9))
model.to(device)

for epoch in range(50):
    model.train()
    run_epoch(data_gen(V, 10, 5670,s1,t1), model, 
              SimpleLossCompute(model.generator, criterion, model_opt))
    model.eval()
    l1,a1,sa1 = run_epoch(data_gen(V, 10,737 ,s1,t1), model, 
              SimpleLossCompute(model.generator, criterion, model_opt))
    l,a2,sa2 = run_epoch(data_gen(V, 10, 737,s9,t9), model, 
                    SimpleLossCompute(model.generator, criterion, None))
    #l2,a3,sa3 = run_epoch(data_gen(V, 10, 737,s4,t4), model, 
                    #SimpleLossCompute(model.generator, criterion, None))
    torch.save(model,'simpletod/dialog_NLP.pt') #saving the model in specified location
    #loss_plot2.append(l2)
    #epochs.append(epoch)
    #loss_plot.append(l)
    #loss_plot1.append(l1)
    #accuracy.append(s)
    file1 = open("loss_file5.txt", "a")
    file1.write("Epoch %d \n" % (epoch+1)) # append mode 
    file1.write("Train loss %f \n" % l1) 
    file1.write("Validation loss %f \n" % l)
    #file1.write("Test loss %f \n" % l2) 
    file1.write("Joint accuracy train %f \n" % a1)
    file1.write("Joint accuracy val %f \n" % a2)
    #file1.write("Joint accuracy test %f \n" % a3)
    file1.write("Slot accuracy train %f \n" % sa1)
    file1.write("Slot accuracy val %f \n" % sa2)
    #file1.write("Slot accuracy test %f \n" % sa3) 
    file1.close()
    print(epoch+1)
    print(l)
    print(l1)
```
